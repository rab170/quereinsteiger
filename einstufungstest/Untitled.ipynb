```python
import ggplot
import numpy as np
from numpy import pi
import pandas as pd
```

```python
x = np.linspace(0, 1, num=100) 
y = map(lambda x: np.sin(2*pi*x), x) + np.random.normal(-0.1, 0.1 , 100) 
df = pd.DataFrame({'x':x, 'y':y})
ggplot.ggplot(aes(x='x', y='y'), data=df) + ggplot.geom_point()
```

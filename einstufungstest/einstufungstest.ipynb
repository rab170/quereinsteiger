```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for path in glob.glob('data/*.csv'):
    df = pd.read_csv(path)
    counts = df.groupby('level').count().reset_index()
    ax = counts.plot(kind='bar')
    ax.set(xlabel='Language Level', ylabel='Student Count', title=os.path.basename(path)[:-4], xticklabels=counts.level)
    ax.legend().set_visible(False)
    plt.show()
```
